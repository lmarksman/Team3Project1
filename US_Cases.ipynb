```python
import pandas as pd
```

```python

file1 = "Resources/United_state_death_rate.csv"

file2 = "Population.csv"

```

```python
df1 = pd.read_csv(file1)
df1.head()
```

```python
df2 = pd.read_csv(file2)
df2.head()
```

```python
df1 = df1.rename(columns={"state":"State", "submission_date": "Date", "new_case": "New_Cases"})

df2 = df2.rename(columns={"2019 Population":"Total Population", "State":"US_States", "Code":"State"})
```

```python
df_combined = pd.merge(df1,df2, on = "State")
df_combined.head()
```

```python
df_combined=df_combined[["Date", "Week", "State", "New_Cases", "Total Population"]]
df_combined
```

```python
df_combined=df_combined.replace({'Week': {53.0: 0, 52.0: -1.0, 51.0:-2.0, 50.0:-3.0, 49.0:-4.0}})
df_combined
```

```python
summary_df = pd.DataFrame({
        "Date": df_combined["Date"],
        "Week": df_combined["Week"].map("{0:,.0f}".format),
        "New_Cases": df_combined["New_Cases"],
        "State": df_combined["State"]
       
    
})
summary_df.head()
```
